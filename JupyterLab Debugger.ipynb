## Variable Tracing
We saw on the previous page that variables can have their values updated. Variable tracing is the process of monitoring the value stored in each variable after running each line of code. We could manually trace code by creating a trace table.

Example code:

```python
alpha = 1
beta = 3
alpha = alpha + 1
gamma = alpha + beta
```

Corresponding trace table:

|Code line | alpha | beta | gamma |
|----------|-------|------|-------|
|<code>alpha = 1</code> |	1  | - | - |
|<code>beta = 3</code> |	1 | 3 |-|
|<code>alpha = alpha + 1</code> |2 |3 |- |
|<code>gamma = alpha + beta</code> |	2 |3 |5 |

While manual tracing like this is a good learning exercise, it doesn't scale particularly well for very large pieces of code. Thankfully, JupyterLab has a debugging tool that make it easy to track the state of variables.

***
## The JupyterLab Debugger
The debugger will initially be closed. To open the debugger, you need to click on the bug icon near the top of the screen. The bug icon will turn orange when it is on.



The debugging tool will give you some added functionality while programming.

* It provides a **variable explorer**, which allows you to see all the variables you have created and what values are currently assigned to them.
* You can now add **breakpoints** to your code, which are points where the code will pause. These are useful if you have semantic errors in your code and you are trying to diagnose where they occur.
* The **callstack** will allow you to run a code cell one line at a time, allowing you to trace variables through JupyterLab.
