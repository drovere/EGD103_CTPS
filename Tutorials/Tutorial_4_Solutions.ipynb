# While Loops
While loops are used for tasks where the number of iterations required is unknown. While loops continue to iterate while a certain condition is True, so they will iterate until that condition becomes False. They are commonly used for numerical methods, where we require convergence or a certain level of accuracy before exiting from the algorithm.

<a name="count_digits"></a>
## Count Digits

```python
# Given any positive integer, write a function to calculate how many digits in that number.
# Hints:
# * Dividing a number by 10 will reduce the integer part by 1 digit
# * Use a WHILE loop to keep count of how many times the number can be divided by 10.
# * Use the int() function to determine the integer part of a number

def count_digits(integer):
    count = 0
    while integer >= 1:
        count = count + 1
        integer = integer / 10
    return count
```

```python
count_digits(123456789) # answer should be  9
```

```python
count_digits(10101010) # Test answer should be 8
```

```python
count_digits(7) # Test answer should be 1
```

```python
count_digits(724) # Test answer should be 3
```

## Convert to Binary

```python
# Write a function that expects a positive integer and returns a 
# list of binary numbers (0 or 1) that correspond to the binary representation of that integer.
# NB there are built-in functions to convert from binary to
# decimal (int) and vice versa (bin), but instead use the algorithm
# implicit in this example to create your own solution:
#
# 23 to binary:
# / 2 = 11 r 1
# / 2 = 5  r 1
# / 2 = 2  r 1
# / 2 = 1  r 0
# / 2 = 0  r 1
# The binary number can be determined from the series of remainder values,
# from last to first ie [1,0,1,1,1]

# Hint: Remainder is calculated using the % modulo operator.

def convert_to_binary(number):
    binary = []
    while number > 0:
        # find remainder and insert as first digit
        remainder = number % 2
        binary.insert(0, remainder)

        # divide number by 2 and round down
        number = number // 2
    return binary
```

```python
convert_to_binary(23) # answer should be [1, 0, 1, 1, 1]
```

```python
convert_to_binary(7) # answer should be [1, 1, 1]
```

```python
convert_to_binary(2) # answer should be [1, 0]
```

## Nested Loops
Nested loops occur when we place a loop in the body of another loop. They provide a way to iterate through all combinations of pairs that can be created by combining two sequences. A common use of nested loops is to iterate throuh 2-dimensional data structures (eg. matrices).

### Creating an Account
In the lecture we worked through an example piece of code that simulated someone logging in to their account. A dictionary was used to store all the login details. User input and formatted strings were used to communicate back and forth with the person logging in. In this activity we would like you to create a function that simulates someone creating their account.

You should create a function that firstly asks the user for their username. If the username doesn't exist in the login_details dictionary, they can then be asked to enter a password. If the username does exist, they should be given a message that says that the username is already taken. It should then prompt them once again to enter a username. Because we don't know how many times this will happen, you will need to program this process with a while loop.

When the user is asked to enter a password, your code must check that it is a strong password. It needs to have 10+ characters and must contain at least one lower case letter, one upper case letter and one number. You should look through the different string methods that are available to work out how to do these checks. If the password is strong, then the code should store the username as a key of the dictionary and the password as its corresponding value. Otherwise, the code should tell the user that it didn't follow the rules of a strong password and should ask them again to enter a password. Again, you will need a while loop for this, since you don't know how many attempts the person will need before their password in strong.

```python
# creating an empty dictionary to store login details
login_details = {}
```

```python
def create_account():
    # creating username
    unique = False
    while not unique:
        username = input('What is your username?')
        if username in login_details:
            print(f'The username "{username}" is already taken. Please select a different username.')
        else:
            unique = True

    # creating password
    valid_password = False
    while not valid_password:
        password = input('What is your password?')
        # check for length requirement
        if len(password) < 10:
            print('Password must have at least 10 characters')
            continue

        # check for character requirements
        is_capital = False
        is_lower = False
        is_number = False
        for char in password:
            if char.islower():
                is_lower = True
            elif char.isupper():
                is_capital = True
            elif char.isnumeric():
                is_number = True
            if is_lower and is_capital and is_number:
                valid_password = True
                break
        
            
```
