## Programming for Engineers
Learning the fundamentals of programming is an essential skill for engineers, regardless of discipline. Much like machinery can be used to amplify physical strength, programming can be used to amplify mental strength. Programming can be used to automate mundane tasks involving PDF files, Excel spreadsheets and retrieving online information, allowing you to use your time more efficiently (see https://automatetheboringstuff.com/ for a good source on this). Programming can also be used to perform powerful tasks that would otherwise be impossible, such as creating a dynamic simulation of an engineering system, or creating a machine learning model that forecasts how an engineering system will behave in the future. More immediately, you will find programming as a useful tool when completing your engineering studies. You might use programming to automate all the calculations for an engineering assignment, or to create compelling data visuals to include in a report.

Learning to program also offers a lot of indirect benefits that will help you in your career as an engineer including:

* Being able to follow an algorithm (solution process). 
* Being precise and having a high attention to detail.
* Being able to creatively solve complex problems.
* Being able to clearly and efficiently communicate a solution to a problem.

## Data for Engineers
In the current day and age, the amount of data being collected is exponentially increasing. Extremely large datasets are being used by engineers to make informed decisions about complex engineering systems. For example, based on historical traffic data we could identify which areas of a road network require upgrading. What was once a task exclusively given to data analysts, engineers are now expected to be able to wrangle, summarise and visualise large datasets given their contextual understanding of the data. This unit looks to develop your ability apply descriptive statistics (ie. describe and summarise what happened in the collected data) to large engineering datasets. This will serve as a solid foundation for more advanced statistics principles such as inference and prediction that you will learn later in the Bachelor of Engineering degree.
