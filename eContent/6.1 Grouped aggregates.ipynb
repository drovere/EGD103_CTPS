## Grouped Aggregates
Last week you learned how to compute statistical aggregates using Pandas methods. In many cases, we wish to perform an aggregate for each category in a set of data. For example, suppose we wanted to find the average pollution level for each category in the wind direction column. We can simply achieve this with one line of code like so:

```python
import pandas as pd
pollution_data = pd.read_csv('LSTM-Multivariate_pollution.csv', index_col = 'date', parse_dates = True, dayfirst = True)
pollution_data.groupby('wnd_dir')['pollution'].mean()
```

This line of code says to:

1. Split the data into groups based on the wnd_dir column.
2. Select the 'pollution' column of each of those groups.
3. Apply the mean aggregate.

The sequencing here is very important. Generally for a grouped aggregate you want to group, then select, then aggregate.
