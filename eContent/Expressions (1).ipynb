## Mathematical Expressions
This week we will focus on evaluating mathematical expressions just like a calculator! To evaluate mathematical expressions, we can utilise operators from the table below. Each of these operators requires a value to its left and right. 

| Operator | Description |
| ----------- | ----------- |
| + | Addition |
| - | Subtraction |
| * | Multiplication |
| / | Division |
| ** | Exponentiation |
| // | Floored division |
| % | Remainder/modulus |

**Example:** A stopwatch shows a time of 12 minutes and 48 seconds. Convert the time to seconds.

```python
12 * 60 + 48
```

**Example:** A stopwatch shows a time 344 seconds. Convert the time to minutes and seconds.

```python
# minutes
344 // 60
```

```python
# seconds
344 % 60
```

## Operator Precedence
When multiple operators are used in a single expression, Python has rules about what order those operations should be performed in. For mathematical operations, the order Python uses follows your standard BOMDAS rules from mathematics.

1. Brackets
2. Exponentiation (**)
3. Multiplication and division (*, /, //, %)
4. Addition and subtraction (+, -)

Remember to use brackets when needed to control the order of operations. More details can be found here: https://docs.python.org/3/reference/expressions.html#evaluation-order

**Example:** Evaluate the mathematical expression $\cfrac{3^2 + 1}{5}$. 

*Note:* The default order of operations will perform division before addition. We need brackets to ensure addition happens before division.

```python
(3 ** 2 + 1) / 5
```
