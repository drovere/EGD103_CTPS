## Plot customisation
The pandas plot method contains many optional arguments that you can use to customise a plot. Some of the more notable ones are:

* <code>figsize</code>: a two-element tuple that specifies the width and height of the figure
* <code>title</code>: a string that is added as the title for a plot
* <code>xlabel</code>: a string that is used as the x label for the plot
* <code>ylabel</code>: a string that is used as the y label for the plot
* <code>legend</code>: a Boolean specifying whether to add a legend to the plot
* <code>xlim</code>: a tuple that specifies the x-axis limits
* <code>ylim</code>: a tuple that specifies the y-axis limits
* <code>col</code>: specify the colour of the line

You should use these to ensure the plot is as clear and informative as possible. For example, consider the plot below that shows how pollution levels vary over time throughout the data.

```python
import pandas as pd
pollution_data = pd.read_csv('LSTM-Multivariate_pollution.csv', index_col = 'date', parse_dates = True, dayfirst = True)
pollution_data.plot(y = 'pollution')
```

This plot is easy to produce, but needs improvement in order to be an effective data visual. The graph needs a title and a y-axis label to effectively communicate what information is represented. No legend is needed since there is only one line in the graph. The information here is too dense, making it hard to interpret anything. One solution is to zoom in on a particular month rather than looking at roughly 5 years of data.

Below is some sample code that applies all these customisations using optional arguments from the pandas plot method. If using many arguments like this, spreading them over multiple lines can be useful to improve readability.

```python
pollution_data.plot(y = 'pollution', 
                    legend = False, 
                    ylabel = 'pollution level', 
                    title = 'Pollution levels for January, 2011', 
                    xlim = ('2011-01-01 00:00:00', '2011-02-01 00:00:00'), 
                    ylim = (0,300))
```
