## Data visualisation
Data visualisation is a powerful tool to inform people about information contained in the dataset. A good data visual should:

* Contain enough information to be self-sufficient (eg. axis labels, legend). One shouldn't have to read through the body of a report to have to understand what is being represented in a graph.
* Avoid unnecessary clutter. The viewers eyes should only be drawn to meaningful information.
* Answer some kind of question about the data. We don't create visuals for the sake of it, we do it to answer important questions.

There are many different data visualisation libraries in Python. For this subject, we will mainly focus on what pandas offers. We will also briefly cover matplotlib and seaborn and discuss when it is advantageous to use them.

***
## Plotting with pandas
You can create data visuals with pandas using the plot method - https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html. 

This method can be called on either a series or a dataframe. The most common arguments for the method are:

* <code>x</code>: a label specifying what column to place on the x-axis. Defaults to the index.
* <code>y</code>: a label or list of labels specifying what column(s) to place on the y-axis. Defaults to all columns.
* <code>kind</code>: a string specifying the type of plot to create. Defaults to a line graph.

Example: visualising relationship between temperature and pressure in the pollution dataset.

```python
import pandas as pd
pollution_data = pd.read_csv('LSTM-Multivariate_pollution.csv', index_col = 'date', parse_dates = True, dayfirst = True)
pollution_data.plot(x = 'press', y = 'temp', kind = 'scatter')
```
