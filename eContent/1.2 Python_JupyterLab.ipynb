## Introduction to Python
In this unit, you will be learning how to program in Python. Python is an ideal first programming language to learn as an engineer because:

* It is a simple and easy-to-read language, making it ideal for beginners to programming.
* There are libraries available that make engineering-specific tasks such as data processing, numerical modelling and machine learning easy to perform.
* It is the most popular programming language in the world (Open Universities Australia, 2024), meaning you will have access to an almost unlimited amount of learning resources.
* It is open source, meaning it is free to use!

## Introduction to Python Programming in JupyterLab
In this unit, you will be using JupyterLab as your Python development environment. JupyterLab allows you to create notebooks that allow you to blend Python code with word processing features. QUT hosts an online JupyterLab environment through the URL https://jupyter.eres.qut.edu.au/, meaning you do not need to install anything on your computer.

When you program in a JupyterLab notebook, you can choose between 3 types of cells:

* **Code cell:** these will allow you to enter Python code which can be executed. A code cell will display the value of the last expression it contains.
* **Markdown cell:** these will allow you to create sections of formatted text using the Markdown language. You will only need to use Markdown once you start working on assignments. 
* **Raw cell:** Displays raw, unformatted text. We will not use these in this unit.
