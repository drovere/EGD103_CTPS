## Data Processing for Engineers
Data processing has become an essential skill for engineers. With increases in computing power and advances in machine learning, data driven solutions to engineering problems are becoming popular across all disciplines. Data allows us to explore the behaviour of complex engineering systems that can't be deterministically modelled (eg. transportation networks, electricity grids, processing plants). It can also be used to conditionally monitor equipment or to validate simulations/models.

Data processing in EGD103 will consist of the following concepts:

* **Data wrangling**: This is the process of taking some raw data and converting it to a useable form. 
* **Summarising data**: This is the process of taking data and summarising it with statistics.
* **Visualising data**: Data visualisation can clearly and concisely describe trends and relationships in the data.

The content will primarily focus on programming skills for data processing. The only mathematical knowledge required will be an understanding of basic statistical aggregates (eg. mean, median, variance). More advanced statistical concepts such as inference, modelling and prediction are covered later in the Bachelor program.
