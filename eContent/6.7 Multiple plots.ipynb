## Plotting multiple columns
The pandas plot method will allow you to supply a list for the y-variable, allowing you to have multiple plots in the same figure. It will automatically add a legend to the plot for you.

```python
import pandas as pd
pollution_data = pd.read_csv('LSTM-Multivariate_pollution.csv', index_col = 'date', parse_dates = True, dayfirst = True)
pollution_data.plot(y = ['snow', 'rain'])
```

In some cases you would prefer to have the columns plotted in different axes. In this case you can set the <code>subplots</code> argument to <code>True</code>.

```python
pollution_data.plot(y = ['snow', 'rain'], subplots = True)
```

If you want different x-axes for your plots, or different types of plots altogether, then you are best off using matplotlib to provide lower level instruction. The subplots function allows you to create a figure with multiple axes - the arguments specify the number of rows and columns for the subplots. When creating a pandas plot, you can specify which set of axes to plot onto using the ax argument.

For more advanced cases of multiple plots you will want to use matplotlib to provide lower level instruction. This will be covered later in the unit.
